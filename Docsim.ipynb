```python
import pandas as pd
import gensim
from gensim.models.doc2vec import  TaggedDocument
import multiprocessing
import re
import numpy as np
import pathlib
```

```python
pd.set_option('display.max_columns', 5)
pd.set_option('display.max_rows', None) 
pd.set_option('display.max_colwidth', 200) 
df=pd.read_csv(r'data\tagged_plots_movielens1.csv',sep=',',encoding='latin1')
df.count()
docs=[]
docs_all=[]
j=0
df_bkp=df
df=df.dropna()

for row in df.iterrows():
            print(row[1])
        #if         
            docs.append(row[1]["plot"].upper().split())
            print(row[1]["plot"].split())
            docs_all.append(row)
            j += 1
print ("loading done")
word = input("Please enter the string : ")
word=word.upper()
number = int(input("What is number of results you want : "))
```

```python
documents = [TaggedDocument(doc, ['Set_%s'%(i)]) for i, doc in enumerate(docs)]
dm = 0 # 1 for distributed memory(default); 0 for dbow 
cores = multiprocessing.cpu_count()
size = 500
context_window = 2
seed = 42
min_count = 1
alpha = 0.025
max_iter = 2000
# BUILD MODEL
model = gensim.models.doc2vec.Doc2Vec(documents = documents,
dm = dm,
alpha = alpha, # initial learning rate
seed = seed,
cores=cores,
min_count = min_count, # ignore words with freq less than min_count
max_vocab_size = None, 
window = context_window, # the number of words before and after to be used as context
vector_size = size, # is the dimensionality of the feature vector
sample = 1e-4, # ?
negative = 5, # ?
workers = cores, # number of cores
epoch = max_iter # number of iterations (epochs) over the corpus)
)
```

```python
file = pathlib.Path("data\model_moviesim_upd")
if file.exists():
    model = gensim.models.doc2vec.Doc2Vec.load(r'data\model_moviesim_upd')
else:
    model.train(documents,total_examples=model.corpus_count, epochs=1500)
    model.save(r'data\model_moviesim_upd')

new_sentence = word.split()  
op=model.docvecs.most_similar(positive=[model.infer_vector(new_sentence)],topn = number)
#print (op[0].vector)
#print(op)

for output in op:
    (doc_index,simscore)=(int(output[0].strip("Set_")),output[1])
    #print(doc_index)
    print("Top",number ,"most similar movie ids are", docs_all[doc_index][1]["movieId"],"with similarity score of ", simscore)
```
